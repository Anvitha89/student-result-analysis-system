```python
#1. Import Libraries
import numpy as np
import pandas as pd
```

```python
#2. Load Datasets
Student_Marks=pd.read_excel("C:\\Users\\mvr28\\Desktop\\ANVITHA\\PANDAS\\Marks.xlsx",sheet_name="Student_Marks")
Student_Marks_with_Duplicates=pd.read_excel("C:\\Users\\mvr28\\Desktop\\ANVITHA\\PANDAS\\Marks.xlsx",sheet_name="Student_Marks_Dirty")
Student_Details=pd.read_excel("C:\\Users\\mvr28\\Desktop\\ANVITHA\\PANDAS\\Marks.xlsx",sheet_name="Student_Details")
```

```python
#3. Data Exploration
#Dataset Shape
print(Student_Marks.shape)
print(Student_Marks_with_Duplicates.shape)
print(Student_Details.shape)

#Column names
print(list(Student_Marks.columns))
print(list(Student_Marks_with_Duplicates.columns))
print(list(Student_Details.columns))

#Data Types
print(Student_Marks.dtypes)
print(Student_Marks_with_Duplicates.dtypes)
print(Student_Details.dtypes)

#Dataset Information
Student_Marks.info()
Student_Marks_with_Duplicates.info()
Student_Details.info()

#Statistical Summary
print(Student_Marks.describe())
print(Student_Marks_with_Duplicates.describe())
print(Student_Details.describe())
```

```python
#4. Feature Engineering

#Total
Student_Marks["Total"]=Student_Marks["Telugu"]+Student_Marks["English"]+Student_Marks["Maths"]+Student_Marks["Science"]+Student_Marks["Social"]

#Percentage Calculation
Student_Marks["Percentage"]= (Student_Marks["Total"]/500)*100

#Assign Grades
Student_Marks["Grades"]="Pass/Fail"
Student_Marks.loc[Student_Marks["Percentage"]<40, ["Grades"]] = "Fail"
Student_Marks.loc[(Student_Marks["Percentage"]>=40) & (Student_Marks["Percentage"]<=60), ["Grades"]] = "Pass"
Student_Marks.loc[(Student_Marks["Percentage"]>60) & (Student_Marks["Percentage"]<=80), ["Grades"]] = "First Class" 
Student_Marks.loc[Student_Marks["Percentage"]>80, ["Grades"]] = "Distinction"
print(Student_Marks)
```

```python
#5. Student Analysis

#Overall Topper
Highest_Marks= Student_Marks["Total"].max()
Topper_Student= Student_Marks.loc[Student_Marks["Total"]==Highest_Marks, ["Name","Total","Percentage","Grades"]]
print(Topper_Student)

#Bottom Student
Lowest_Marks = Student_Marks["Total"].min()
Bottom_Student = Student_Marks.loc[Student_Marks["Total"]==Lowest_Marks , ["Name","Total","Percentage","Grades"]]
print(Bottom_Student)

#Top 5 Students
Top_Students = Student_Marks.sort_values("Total",ascending=False)
Required = Top_Students.head()[["Name","Total","Percentage","Grades"]]
print(Required)

#Pass Students
Pass_Students = Student_Marks.loc[Student_Marks["Grades"]=="Pass" , ["Name","Total","Percentage","Grades"]]
print(Pass_Students)

# Fail Students
Fail_Students = Student_Marks.loc[Student_Marks["Grades"]== "Fail" ,["Name","Total","Percentage","Grades"]]
print(Fail_Students)

# Distinction Students
Distinction_Students = Student_Marks.loc[Student_Marks["Grades"]=="Distinction",["Name","Total","Percentage","Grades"]]
print(Distinction_Students)

#First class Students
First_Class_Students = Student_Marks.loc[Student_Marks["Grades"]=="First Class",["Name","Total","Percentage","Grades"]]
print(First_Class_Students)
```

```python
#6. Subject Analysis

#Telugu Topper
Telugu_Topper = Student_Marks["Telugu"].max()
Topper_of_Telugu = Student_Marks.loc[Student_Marks["Telugu"]==Telugu_Topper, ["Name","Telugu"]]
print(Topper_of_Telugu)

# English Topper
English_Topper = Student_Marks["English"].max()
Topper_of_English = Student_Marks.loc[Student_Marks["English"]==English_Topper, ["Name","English"]]
print(Topper_of_English)

# Maths Topper
Maths_Topper = Student_Marks["Maths"].max()
Topper_of_Maths = Student_Marks.loc[Student_Marks["Maths"]==Maths_Topper, ["Name","Maths"]]
print(Topper_of_Maths)

# Science Topper
Science_Topper = Student_Marks["Science"].max()
Topper_of_Science = Student_Marks.loc[Student_Marks["Science"]==Science_Topper, ["Name","Science"]]
print(Topper_of_Science)

# Social Topper
Social_Topper = Student_Marks["Social"].max()
Topper_of_Social = Student_Marks.loc[Student_Marks["Social"]==Social_Topper, ["Name","Social"]]
print(Topper_of_Social)

#Subject Analysis

#Average Telugu Marks
Avg_Telugu = Student_Marks["Telugu"].mean()
print("Telugu_Average_Marks=", Avg_Telugu)

#Average English Marks
Avg_English = Student_Marks["English"].mean()
print("English_Average_Marks=", Avg_English)

#Average Maths Marks
Avg_Maths = Student_Marks["Maths"].mean()
print("Maths_Average_Marks=", Avg_Maths)

#Average Science Marks
Avg_Science = Student_Marks["Science"].mean()
print("Science_Average_Marks=", Avg_Science)

#Average Social Marks
Avg_Social = Student_Marks["Social"].mean()
print("Social_Average_Marks=", Avg_Social)

#Highest Marks

#Highest Telugu Marks
Highest_Telugu_Marks = Student_Marks["Telugu"].max()
print("Highest Telugu Marks=", Highest_Telugu_Marks)

#Highest_English_Marks
Highest_English_Marks = Student_Marks["English"].max()
print("Highest English Marks=", Highest_English_Marks)

#Highest_Maths_Marks
Highest_Maths_Marks = Student_Marks["Maths"].max()
print("Highest Maths Marks=", Highest_Maths_Marks)

#Highest_Science_Marks
Highest_Science_Marks = Student_Marks["Science"].max()
print("Highest Science Marks=", Highest_Science_Marks)

#Highest_Social_Marks
Highest_Social_Marks = Student_Marks["Social"].max()
print("Highest Social Marks=", Highest_Social_Marks)

#Lowest Marks

#Lowest_Telugu_Marks
Lowest_Telugu_Marks = Student_Marks["Telugu"].min()
print("Lowest Telugu Marks=", Lowest_Telugu_Marks)

#Lowest_English_Marks
Lowest_English_Marks = Student_Marks["English"].min()
print("Lowest English Marks=", Lowest_English_Marks)

#Lowest_Maths_Marks
Lowest_Maths_Marks = Student_Marks["Maths"].min()
print("Lowest Maths Marks=", Lowest_Maths_Marks)

#Lowest_Science_Marks
Lowest_Science_Marks = Student_Marks["Science"].min()
print("Lowest Science Marks=", Lowest_Science_Marks)

#Lowest_Social_Marks
Lowest_Social_Marks = Student_Marks["Social"].min()
print("Lowest Social Marks=", Lowest_Social_Marks)

#Subject Performance Comparison

#Which subject has the highest average marks?
print("Highest avg marks is Maths", Avg_Maths)

#Which subject has the lowest average marks?
print("Lowest avg marks is Social", Avg_Social)
```

```python
#7. Branch Analysis

#Count students in each branch.
Count_Branch = Student_Marks.groupby(Student_Marks["Branch"])["Name"].count()
print(Count_Branch)

#Average Percentage by Branch
Avg_Perc = Student_Marks.groupby(Student_Marks["Branch"])["Percentage"].mean()
print(Avg_Perc)

#Average Telugu Marks by Branch
Avg_Telugu = Student_Marks.groupby(Student_Marks["Branch"])["Telugu"].mean()
print(Avg_Telugu)

#Average English Marks by Branch
Avg_English = Student_Marks.groupby(Student_Marks["Branch"])["English"].mean()
print(Avg_English)

#Average Maths Marks by Branch
Avg_Maths = Student_Marks.groupby(Student_Marks["Branch"])["Maths"].mean()
print(Avg_Maths)

#Average Science Marks by Branch
Avg_Science = Student_Marks.groupby(Student_Marks["Branch"])["Science"].mean()
print(Avg_Science)

#Average Social Marks by Branch
Avg_Social = Student_Marks.groupby(Student_Marks["Branch"])["Social"].mean()
print(Avg_Social)
```

```python
#7. Branch Analysis

#Count Pass Students in each Branch
Pass_Students = Student_Marks.loc[Student_Marks["Grades"]=="Pass"]
Branch_Count = Pass_Students.groupby(Student_Marks["Branch"])["Name"].count()
print(Branch_Count)

# Count Fail Students in each Branch
Fail_Students = Student_Marks.loc[Student_Marks["Grades"]=="Fail"]
Branch_Count_Fail = Fail_Students.groupby(Student_Marks["Branch"])["Name"].count()
print(Branch_Count_Fail)

# Count First Class Students in each Branch
First_Class_Students = Student_Marks.loc[Student_Marks["Grades"]=="First Class"]
Branch_Count_First_Class = First_Class_Students.groupby(Student_Marks["Branch"])["Name"].count()
print(Branch_Count_First_Class)

# Count Distinction Students in each Branch
Distinction_Class_Students = Student_Marks.loc[Student_Marks["Grades"]=="Distinction"]
Branch_Count_Distinction = First_Class_Students.groupby(Student_Marks["Branch"])["Name"].count()
print(Branch_Count_Distinction)
```

```python
#8. Data Cleaning

Student_Marks_with_Duplicates=pd.read_excel("C:\\Users\\mvr28\\Desktop\\ANVITHA\\PANDAS\\Marks.xlsx",sheet_name="Student_Marks_Dirty")

#Check missing values
Missing_Values = Student_Marks_with_Duplicates.isnull().sum()
print(Missing_Values)

#Check duplicate rows
Duplicate_rows = Student_Marks_with_Duplicates.duplicated().sum()
print(Duplicate_rows)

#Display duplicate rows
Display_Duplicate_rows = Student_Marks_with_Duplicates[Student_Marks_with_Duplicates.duplicated()]
print(Display_Duplicate_rows)

# Remove duplicate rows
Remove_Duplicate_rows = Student_Marks_with_Duplicates.drop_duplicates()
print(Remove_Duplicate_rows)

a = Remove_Duplicate_rows.duplicated().sum()
print(a)
```

```python
#Merge Datasets
#Merge Student_Marks and Student_Details using Roll_No.
Merged = pd.merge(Student_Marks, Student_Details, on="Roll_No")
print(Merged)

#Check shape
rows_cols = Merged.shape
print(rows_cols)

#Check column names
cols = list(Merged.columns)
print(cols)

#Display only: Name Branch City Total Percentage Grades
Specific = Merged[["Name","Branch","City","Total","Percentage","Grades"]]
print(Specific)

#Students from Vijayawada Display their details.
Vijayawada_Details = Merged.loc[Merged["City"]=="Vijayawada", ["Name","Branch","Telugu","English","Maths","Science","Social","Total","Percentage","Grades","City","Phone"]]
print(Vijayawada_Details)

#Students with Percentage > 80 Display: Name City Percentage
Required = Merged.loc[Merged["Percentage"]>80, ["Name","City","Percentage"]]
print(Required)

#Students from CSE branch Display: Name City Percentage
Cse = Merged.loc[Merged["Branch"]=="CSE", ["Name","Branch","City","Percentage"]] 
print(Cse)

#Average Percentage by City
Avg_City = Merged.groupby(Merged["City"])["Percentage"].mean()
print(Avg_City)
```
